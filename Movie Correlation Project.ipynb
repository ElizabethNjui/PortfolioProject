```python
# Visualizing the correlation matrix

correration_matrix = df.corr(method='pearson')

sns.heatmap(correration_matrix, annot=True)


plt.title('Correlation Matric For Numeric Features')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
# Determine the correlation between Company and Gross
# Converting Company(str) to numeric value to enable for visualization

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
       df_numerized[col_name] = df_numerized[col_name].astype('category')
       df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python

correration_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correration_matrix, annot=True)


plt.title('Correlation Matric For Numeric Features')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
df_numerized.corr()
```

```python
# Organize this so the ones with highest correlation can be identified quickly

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
